##### Celsius to Fahrenheit function

```python
def cel_to_fahr(celsius):
    return (celsius*9/5)+32
print(cel_to_fahr(25))
```

##### Swap without temp (third Variable)


```python
def swap_without_temp(a,b):
    return b,a
result = swap_without_temp(5,10)
print(f"a={result[0]},b={result[1]}")
```

##### Transpose a Matrix

```python
def transpose_matrix(matrix):
    return [list(row) for row in zip(*matrix)]
matrix = [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ]
print("\n".join(",".join(map(str,row)) for row in transpose_matrix(matrix)))
```

##### check the validity of the date

```python
from datetime import datetime
def is_date_valid(val):
    return datetime.strptime(val,"%B %d, %Y %H:%M:%S") if val else False
print(is_date_valid("March 8, 2026 19:16:00"))
```

##### Capitalize a string

```python
def cap_string(s):
    return s[0].upper() + s[1:] if s else ""
print(cap_string("this is python"))
```

#### Find the difference between two days

```python
from datetime import datetime
def day_diff(date1,date2):
    return abs((date2-date1).days)

date1 = datetime(2026,3,8)
date2 = datetime(2026,3,7)
print(day_diff(date1,date2))
```

##### Find the character Frequency in a string

```python
from collections import Counter
s = "hello world"
Counter(s)
```

```python
### with out importing collections
def character_freq(string):
    return {char:string.count(char) for char in set(string)}
print(character_freq("hello world"))
```

##### Reverse a String


```python
def rev_string(s):
    return (s[::-1])
print(rev_string("Linux"))
```

##### Shuffle an array

```python
import random
def shuffle_array(arr):
    shuffled = arr[:]
    random.shuffle(shuffled)
    return shuffled
print(",".join(map(str,shuffle_array([1,2,3,4]))))
```

##### Count total number of True values in an array

```python
def count_true(arr):
    return sum(1 for x in arr if x)
bool_array=[True,True,False,True,False]
print(count_true(bool_array))
```

#### The cap_to_front function moves all captial letters in a word to the fron of the word, preserving their original order, and followed the lower case

```python
def cap_to_front(s):
    return "".join(filter(str.isupper,s)).join(filter(str.islower,s))
print(cap_to_front("MoveCapitalLettersTofront"))
```

#### map() function 
The map() function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.

map(function, iterables)


```python
def map_function_example(n):
    return len(n)
x=map(map_function_example,('apple','orange','grapes'))
print(x)
# convert the map into list for the reabability
print(list(x))
```

##### Count number of words in a sentence

```python
def count_words(sentence):
    return (len(sentence.split()))
print(count_words("Python Programming Language"))
```

#### Count Number of One's(1) in the binary repsenteation of an integer

```python
def count_ones(number):
    return bin(number).count('1')
print(count_ones(20))
```

#### Check if the string is empty

```python
def check_string_empty(string):
    return not string.strip()
print(check_string_empty(""))
```

##### index_multiplier function calculates the sum of all items in an array , where each item is multiplied by its array

```python
def index_multiplier(arr):
    return sum(value*index for index,value in enumerate(arr))
arr=[1,2,3,4]
print(index_multiplier(arr))
```

#### Find max value of an item in an array object

```python
def find_max_value(arr,key):
    return max(item[key] for item in arr)

arr = [
{"name": "Alice", "score": 80},
{"name": "Bob", "score": 95},
{"name": "Charlie", "score": 70}
]

print(find_max_value(arr,"score"))
```

#### Check if a string starts with a specific character

```python
def starts_with_char(string,char):
    return string.lower().startswith(char.lower())

print(starts_with_char("Hello World","H"))
```

#### Check a specific value is it contains in the array

```python
def check_contains(arr,value):
    return value in arr
arr=[1,23,2]
print(check_contains(arr,2))
```

#### Array to CSV

```python
def array_to_csv(arr):
    return ','.join(map(str,arr))
print(array_to_csv([1,2,3]))
```

#### Find the index of an element in an array

```python
def find_index(arr,element):
    return arr.index(element)
print(find_index([1,2,3,4],3))
```

#### Remove an element from an array

```python
def remove_element(arr,element):
    return [x for x in arr if x!=element]
print(remove_element([1,2,3,4,54],4))
```

#### Calculate the `Standard Deviation` of array of numbers

```python
import math
def standard_deviation(arr):
    mean = sum(arr)/len(arr)
    sum_of_squared_difference = sum((x-mean)**2 for x in arr)
    return math.sqrt(sum_of_squared_difference/len(arr))
print(standard_deviation([1,2,3,4,5]))
```

#### Return None items from an array

```python
def remove_none(arr):
    return [item for item in arr if item is not None]

array = [1, None, 2, 3, None, 4, None]
print(remove_none(array))
```
