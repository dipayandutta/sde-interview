#### What is the difference between `str.replace()` and `str.sub()` ?

**`str.replace()`** Simple Substring replacement . No Patter Matching. Faster for basic operations
**`str.sub()`** Regex based replacement . Supports Complex Patterns . Flexible but Slightly heavier

```python
import re 
text = "abc123"

# Replace 123 with XYZ
print(text.replace("123","XYZ"))

# Using Regex
print(re.sub(r"\d+","XYZ",text))
```

#### What is the difference between `shallow copy` and `deep copy` ?

```python
# Shallow Copy -> Copies outer list only
# deep copy -> copies nested objects too 

import copy
l1 = [[1,2],[3,4]]
l2 = copy.copy(l1) # Shallow
l3 = copy.deepcopy(l1) # deepcopy

print(f"Original List: {l1}")
print(f"Shallow Copy List: {l2}")
print(f"Deep Copy List : {l3}")
```

#### Find the First Non Repeating Character in a string

```python
from collections import Counter

def first_unique(s):
    freq = Counter(s) 
    print(f"Frequncey of each Character in the text : {freq}") #Counter({'s': 3, 'w': 1, 'i': 1})
    for ch in s:
        if freq[ch] == 1: # here value/count of w is 1 First Then second one is i
            return ch # As the first w matches 
    return None

print(first_unique("swiss"))
```

```python
# Another approach
from collections import Counter

def first_unique(s):
    freq = Counter(s)
    sorted_result = dict(sorted(freq.items(), key=lambda x: x[1]))
    print(list(sorted_result.keys())[0])

first_unique("swiss")

    
```

#### How do you check if two strings are anagram?

```python
from collections import Counter

def is_anagram(s1,s2):
    return Counter(s1) == Counter(s2)

print(is_anagram("listen","silent"))
```

#### What happens if dictonary keys collide(hash collision)
- Python uses `hash table`
- If two key has same hash -> Collision resolution via open addressing.
- Python compress equality `(__eq__)` to differentiate keys
- Lookup remains average O(1)

```python
class A:
    def __hash__(self):
        return 1

a = A()
b = A()
print(hash(a),hash(b))
print(a==b)
```

#### Invert a dictonary

```python
def invert_dict(d):
    return {v:k for k,v in d.items()}

print(invert_dict({"a":1,"b":2}))
```

#### Rotate a python list by k position


```python
def rotate(lst,k):
    k %= len(lst)
    return lst[-k:] + lst[:-k]

print(rotate([1,2,3,4,5],2))
```
